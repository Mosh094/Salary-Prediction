# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

# Load Data

```python
data = pd.read_csv('Years_of_Experience.csv')
```

# Delete the unnamed column

```python
data = data.drop("Unnamed: 0", axis=1)
```

# CREATE REGRESSION

# Declare the dependent and the independent variables

```python
y = data['Salary']
x1 = data['YearsExperience']
```

### Explore the data with a scatter plot

```python
plt.scatter(x1,y)
plt.xlabel('Years of Experience',fontsize=20)
plt.ylabel('Salary',fontsize=20)
plt.show()
```

# Regression Results

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

# Analysing Coefficients and Results

```python
# Constant  = 25790 and Coefficient of x1 = 9449.9623
# R Squared = 0.957, This means that the model explains almost the entire variability of the model. 
# The result shows a very good Rsquared and the model is good enough that experience explains 96% of variability of salaries. the model shows a good fit
```

# Ploting Regression Line

```python
plt.scatter(x1,y)
yhat = x1*9449.9623+25790
fig = plt.plot(x1,yhat, lw=4, c='orange', label ='regression line')
plt.xlabel('Years of Experience', fontsize = 20)
plt.ylabel('Price', fontsize = 20)
plt.show()
```
